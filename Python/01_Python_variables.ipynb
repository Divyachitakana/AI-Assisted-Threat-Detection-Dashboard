#Python - Variables

## Explanation

Variables are used to store data values in Python. They act as containers for information such as numbers, text, or boolean values. A variable is created when a value is assigned using the `=` operator, and Python automatically determines its data type

## prompt

##Example


```python

age = 20
height = 5.4
name = "Raju"
is_student = True

print(age)
print(height)
print(name)
print(is_student)
```

## Conclusion

Variables allow you to store and manage data efficiently in Python. They make programs easier to read, modify, and reuse.
