# Matplotlib – Scatter Plot

## Explanation

A scatter plot shows the relationship between two numerical variables.

## Prompt

> Explain how to create a scatter plot using Matplotlib with simple examples.

## Example

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,5,3,8,7]

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Conclusion

Scatter plots help identify relationships, trends, and patterns between two variables.
