# Python - Sets

## Explanation

A set is an unordered collection of unique items in Python. Sets do not allow duplicate values and are useful for storing distinct elements. They are created using curly braces `{}` or the `set()` function.

## Prompt

> Explain Python sets in simple language for beginners. Provide an example that demonstrates creating a set, adding an element, removing duplicate values, and printing the updated set.

## Example

```python
# Creating a set

numbers = {1, 2, 3, 4, 4, 5}

# Printing the original set
print("Original set:", numbers)

# Adding a new element
numbers.add(6)

# Printing the updated set
print("Updated set:", numbers)
```

## Conclusion

Sets are useful for storing unique values and automatically remove duplicate elements. They are commonly used for membership testing and performing set operations such as union and intersection.
