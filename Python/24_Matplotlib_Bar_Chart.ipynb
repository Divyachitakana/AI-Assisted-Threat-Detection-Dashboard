# Matplotlib – Bar Chart

## Explanation

Bar charts compare values across different categories.

## Prompt

> Explain how to create a bar chart using Matplotlib.

## Example

```python
import matplotlib.pyplot as plt

subjects=["Python","Java","C++"]
marks=[90,80,85]

plt.bar(subjects,marks)
plt.title("Marks")
plt.show()
```

## Conclusion

Bar charts are useful for comparing values between categories.
