# Python - If-Else

## Explanation

The `if`, `elif`, and `else` statements are used to make decisions in Python. They allow a program to execute different blocks of code based on whether a condition is `True` or `False`.

## Prompt

> Explain Python if, elif, and else statements in simple language for beginners. Provide an example that checks whether a number is positive, negative, or zero using conditional statements.

## Example

```python
# Checking whether a number is positive, negative, or zero

number = int(input("Enter a number: "))

if number > 0:
    print("The number is Positive.")
elif number < 0:
    print("The number is Negative.")
else:
    print("The number is Zero.")
```

## Conclusion

The `if`, `elif`, and `else` statements help a program make decisions based on conditions. They are essential for controlling the flow of a Python program.
