# NumPy – Mathematical Functions

## Explanation

NumPy provides many built-in mathematical functions that work on arrays quickly and efficiently.

## Prompt

> Explain common NumPy mathematical functions like sum, mean, max, min, sqrt, and square with examples.

## Example

```python
import numpy as np

arr = np.array([4,9,16,25])

print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))
print("Square Root:", np.sqrt(arr))
print("Square:", np.square(arr))
```

## Conclusion

NumPy includes many mathematical functions that make calculations on arrays simple and efficient.
