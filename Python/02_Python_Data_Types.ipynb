# Python - Data Types

## Explanation

Data types define the type of value a variable can store in Python. Python has several built-in data types, including integers, floating-point numbers, strings, booleans, lists, tuples, dictionaries, and sets. Python automatically identifies the data type based on the assigned value.

## Prompt

> Explain Python data types in simple language for beginners. Provide an example demonstrating integer, float, string, boolean, list, tuple, dictionary, and set data types, then print each value along with its data type.

## Example

```python
# Different Python data types

age = 20                    # Integer
height = 5.8                # Float
name = "Raju"               # String
is_student = True           # Boolean
colors = ["Red", "Blue"]    # List
fruits = ("Apple", "Mango") # Tuple
student = {"name": "Raju", "age": 20}  # Dictionary
numbers = {1, 2, 3, 4}      # Set

# Printing values and their data types

print(age, type(age))
print(height, type(height))
print(name, type(name))
print(is_student, type(is_student))
print(colors, type(colors))
print(fruits, type(fruits))
print(student, type(student))
print(numbers, type(numbers))
```



## Conclusion

Python provides different built-in data types to store various kinds of data. Choosing the appropriate data type helps organize information and makes programs easier to understand and maintain.
