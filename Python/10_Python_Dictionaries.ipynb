# Python - Dictionaries

## Explanation

A dictionary is a collection of data stored as key-value pairs in Python. Dictionaries are mutable, meaning their contents can be modified after creation. They are created using curly braces `{}` and allow fast access to values using their keys.

## Prompt

> Explain Python dictionaries in simple language for beginners. Provide an example that demonstrates creating a dictionary, accessing values, adding a new key-value pair, and printing the updated dictionary.

## Example

```python
# Creating a dictionary

student = {
    "name": "Raju",
    "age": 20,
    "course": "Python"
}

# Accessing a value

print("Name:", student["name"])

# Adding a new key-value pair

student["city"] = "Hyderabad"

# Printing the updated dictionary

print("Updated dictionary:", student)
```

## Conclusion

Dictionaries are useful for storing data in key-value pairs. They make it easy to organize, access, and update information efficiently using unique keys.
