# Matplotlib – Pie Chart

## Explanation

A pie chart shows the proportion of each category as parts of a whole.

## Prompt

> Explain how to create a pie chart using Matplotlib.

## Example

```python
import matplotlib.pyplot as plt

sizes=[40,30,20,10]
labels=["Python","Java","C++","JavaScript"]

plt.pie(sizes,labels=labels,autopct="%1.1f%%")
plt.title("Programming Languages")
plt.show()
```

## Conclusion

Pie charts help visualize percentage distributions effectively.
