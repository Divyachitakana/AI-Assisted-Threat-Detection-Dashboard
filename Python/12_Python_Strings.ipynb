# Python - Strings

## Explanation

A string is a sequence of characters used to store and manipulate text in Python. Strings are enclosed in single quotes (`' '`), double quotes (`" "`), or triple quotes (`''' '''` or `""" """`). Python provides many built-in methods to work with strings, such as changing case, finding text, and replacing characters.

## Prompt

> Explain Python strings in simple language for beginners. Provide an example that demonstrates creating a string, accessing characters, converting the string to uppercase, and printing the results.

## Example

```python
# Creating a string

message = "Hello, Python!"

# Accessing characters

print("First character:", message[0])

# Converting to uppercase

print("Uppercase:", message.upper())

# Printing the original string

print("Original string:", message)
```

## Conclusion

Strings are used to store and manipulate text in Python. They provide many useful methods that make it easy to process and format textual data.
