# Python - Input and Output

## Explanation

Input and output are used to interact with the user in a Python program. The `input()` function is used to receive input from the user, while the `print()` function is used to display information on the screen.

## Prompt

> Explain Python input and output in simple language for beginners. Provide an example using the `input()` function to accept user input and the `print()` function to display the entered value.

## Example

```python
# Taking input from the user

name = input("Enter your name: ")

# Displaying the input

print("Hello,", name + "!")
```

## Conclusion

The `input()` function allows a program to receive information from the user, while the `print()` function displays the output. Together, they make Python programs interactive.
