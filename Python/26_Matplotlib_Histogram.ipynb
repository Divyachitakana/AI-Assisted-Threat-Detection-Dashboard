# Matplotlib – Histogram

## Explanation

A histogram shows the frequency distribution of numerical data.

## Prompt

> Explain how to create a histogram using Matplotlib.

## Example

```python
import matplotlib.pyplot as plt

marks=[60,65,70,75,80,85,90,95,100]

plt.hist(marks,bins=5)
plt.title("Histogram")
plt.show()
```

## Conclusion

Histograms help understand how data values are distributed.
