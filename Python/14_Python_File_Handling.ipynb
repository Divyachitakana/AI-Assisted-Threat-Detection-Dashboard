# Python - File Handling

## Explanation

File handling in Python allows you to create, read, write, and append data to files. Python provides the built-in `open()` function to work with files in different modes, such as read (`"r"`), write (`"w"`), and append (`"a"`). It is recommended to use the `with` statement, as it automatically closes the file after use.

## Prompt

> Explain Python file handling in simple language for beginners. Provide an example that demonstrates creating a text file, writing data to it, reading the contents, and displaying the result using the `with` statement.

## Example

```python
# Writing to a file

with open("sample.txt", "w") as file:
    file.write("Welcome to Python File Handling!")

# Reading from the file

with open("sample.txt", "r") as file:
    content = file.read()

print(content)
```

## Conclusion

File handling allows Python programs to store and retrieve data from files. Using the `with` statement ensures that files are properly closed after they are used, making the code safer and easier to manage.
