# Python - For Loop

## Explanation

A `for` loop is used to iterate over a sequence, such as a list, tuple, string, or range of numbers. It executes a block of code repeatedly for each item in the sequence, making it useful for performing repetitive tasks.

## Prompt

> Explain the Python `for` loop in simple language for beginners. Provide an example using the `range()` function to print numbers from 1 to 5.

## Example

```python
# Printing numbers from 1 to 5 using a for loop

for i in range(1, 6):
    print(i)
```

## Conclusion

The `for` loop is an efficient way to repeat a block of code for each item in a sequence. It is commonly used to process collections of data and automate repetitive tasks.
