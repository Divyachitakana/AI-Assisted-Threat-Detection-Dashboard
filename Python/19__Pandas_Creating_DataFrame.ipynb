# Pandas – Creating a DataFrame

## Explanation

A DataFrame is a two-dimensional table with rows and columns. It is one of the most important data structures in Pandas.

## Prompt

> Explain how to create a Pandas DataFrame from a dictionary using beginner-friendly examples.

## Example

```python
import pandas as pd

data = {
    "Name":["Alice","Bob","Charlie"],
    "Age":[23,25,22],
    "City":["Delhi","Mumbai","Pune"]
}

df = pd.DataFrame(data)
print(df)
```

## Conclusion

A DataFrame stores data in rows and columns, making it easy to organize and analyze datasets.
