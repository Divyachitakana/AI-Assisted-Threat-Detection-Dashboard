# Python - Lists

## Explanation

A list is an ordered and mutable collection in Python that can store multiple items in a single variable. Lists allow duplicate values and can contain different data types. They are created using square brackets `[]`.

## Prompt

> Explain Python lists in simple language for beginners. Provide an example that demonstrates creating a list, accessing elements, adding a new element, and printing the updated list.

## Example

```python
# Creating a list

fruits = ["Apple", "Banana", "Mango"]

# Accessing the first element

print("First fruit:", fruits[0])

# Adding a new element

fruits.append("Orange")

# Printing the updated list

print("Updated list:", fruits)
```

## Conclusion

Lists are one of the most commonly used data structures in Python. They make it easy to store, access, and modify multiple values in a single variable.
