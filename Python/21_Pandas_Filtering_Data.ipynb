# Pandas – Filtering Data

## Explanation

Filtering allows us to select only the rows that satisfy a specific condition.

## Prompt

> Explain how to filter rows in a Pandas DataFrame using conditions.

## Example

```python
import pandas as pd

data = {
    "Name":["Alice","Bob","Charlie"],
    "Age":[23,25,22]
}

df = pd.DataFrame(data)

print(df[df["Age"]>22])
```

## Conclusion

Filtering helps extract only the required information from a dataset.
