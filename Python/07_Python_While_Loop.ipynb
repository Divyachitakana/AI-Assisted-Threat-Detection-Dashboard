# Python - While Loop

## Explanation

A `while` loop is used to repeatedly execute a block of code as long as a specified condition is `True`. It is useful when the number of iterations is not known in advance. The loop stops automatically when the condition becomes `False`.

## Prompt

> Explain the Python `while` loop in simple language for beginners. Provide an example that prints numbers from 1 to 5 using a `while` loop.

## Example

```python
# Printing numbers from 1 to 5 using a while loop

count = 1

while count <= 5:
    print(count)
    count += 1
```

## Conclusion

The `while` loop is useful for repeating a block of code until a condition becomes `False`. It is commonly used when the number of iterations is not predetermined.
