# Python - Tuples

## Explanation

A tuple is an ordered and immutable collection in Python used to store multiple items in a single variable. Unlike lists, tuples cannot be modified after they are created. Tuples are created using parentheses `()` and can contain different data types.

## Prompt

> Explain Python tuples in simple language for beginners. Provide an example that demonstrates creating a tuple, accessing elements, and printing the tuple.

## Example

```python
# Creating a tuple

fruits = ("Apple", "Banana", "Mango")

# Accessing elements

print("First fruit:", fruits[0])
print("Second fruit:", fruits[1])

# Printing the entire tuple

print("Tuple:", fruits)
```

## Conclusion

Tuples are useful for storing collections of data that should not be changed. They provide a simple and efficient way to keep related values together.
