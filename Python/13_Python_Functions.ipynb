# Python - Functions

## Explanation

A function is a reusable block of code that performs a specific task. Functions help organize programs, reduce code duplication, and improve readability. In Python, functions are defined using the `def` keyword and can accept parameters and return values.

## Prompt

> Explain Python functions in simple language for beginners. Provide an example that demonstrates defining a function, passing parameters, calling the function, and displaying the result.

## Example

```python
# Defining a function

def greet(name):
    print("Hello,", name + "!")

# Calling the function

greet("Raju")
greet("Priya")
```

## Conclusion

Functions make Python programs more organized and reusable. They allow you to write code once and use it multiple times, making programs easier to maintain.
