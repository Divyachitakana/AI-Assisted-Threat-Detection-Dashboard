# Python - Operators

## Explanation

Operators are special symbols in Python used to perform operations on variables and values. Python provides different types of operators, such as arithmetic, comparison, logical, assignment, and membership operators, to perform various tasks in a program.

## Prompt

> Explain Python operators in simple language for beginners. Provide examples of arithmetic, comparison, logical, assignment, and membership operators with clear and easy-to-understand code.

## Example

```python
# Arithmetic Operators

a = 10
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)

# Comparison Operators

print("Is a greater than b?", a > b)
print("Is a equal to b?", a == b)

# Logical Operators

x = True
y = False

print("x and y:", x and y)
print("x or y:", x or y)

# Assignment Operator

c = 10
c += 5
print("Value of c:", c)

# Membership Operator

fruits = ["Apple", "Banana", "Mango"]

print("Apple" in fruits)
print("Orange" not in fruits)
```

## Conclusion

Operators are essential in Python for performing calculations, making comparisons, checking conditions, assigning values, and working with collections. Understanding operators is fundamental to writing effective Python programs.
