# Python - Exception Handling

## Explanation

Exception handling is used to manage errors that occur during the execution of a Python program. It prevents the program from crashing by handling exceptions using the `try`, `except`, and `finally` blocks. This makes programs more reliable and user-friendly.

## Prompt

> Explain Python exception handling in simple language for beginners. Provide an example that demonstrates using `try`, `except`, and `finally` to handle a division by zero error.

## Example

```python
# Handling an exception

try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except ValueError:
    print("Error: Please enter a valid number.")

finally:
    print("Program execution completed.")
```

## Conclusion

Exception handling helps prevent programs from crashing when errors occur. Using `try`, `except`, and `finally` allows you to handle unexpected situations gracefully and improve the reliability of your Python programs.
