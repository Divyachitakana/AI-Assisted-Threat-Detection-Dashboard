# Matplotlib – Line Chart

## Explanation

A line chart is used to show trends or changes over time.

## Prompt

> Explain how to create a simple line chart using Matplotlib.

## Example

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Conclusion

Line charts help visualize trends and changes in data.
