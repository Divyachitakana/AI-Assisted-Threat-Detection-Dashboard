# NumPy – Array Operations

## Explanation

NumPy allows us to perform operations on entire arrays without using loops. This makes the code shorter, faster, and easier to read.

## Prompt

> Explain basic NumPy array operations like addition, subtraction, multiplication, division, and indexing with simple examples.

## Example

```python
import numpy as np

a = np.array([10,20,30])
b = np.array([1,2,3])

print("Addition:", a+b)
print("Subtraction:", a-b)
print("Multiplication:", a*b)
print("Division:", a/b)

print("First Element:", a[0])
print("Last Element:", a[-1])
```

## Conclusion

NumPy performs array operations efficiently without loops, making data processing much faster than regular Python lists.
