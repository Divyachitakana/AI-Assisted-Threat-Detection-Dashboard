# Pandas – Sorting Data

## Explanation

Sorting arranges data in ascending or descending order.

## Prompt

> Explain how to sort data in a Pandas DataFrame using one or more columns.

## Example

```python
import pandas as pd

data = {
    "Name":["Alice","Bob","Charlie"],
    "Age":[23,25,22]
}

df = pd.DataFrame(data)

print(df.sort_values(by="Age"))
```

## Conclusion

Sorting makes datasets easier to understand and analyze.
