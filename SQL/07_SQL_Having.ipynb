# SQL – HAVING

## Explanation

The `HAVING` clause is used to filter grouped data after the `GROUP BY` clause has been applied. While the `WHERE` clause filters individual rows before grouping, the `HAVING` clause filters the groups created by `GROUP BY`.

In this notebook, we will learn how to use the `HAVING` clause with aggregate functions to filter grouped results.

## Prompt

> Explain the SQL HAVING clause in simple language for beginners. Show how to filter grouped data using aggregate functions with practical examples.

## Example

```python
-- Show cities that have more than one student

SELECT City, COUNT(*) AS TotalStudents
FROM Students
GROUP BY City
HAVING COUNT(*) > 1;

-- Show cities where the average age is greater than 21

SELECT City, AVG(Age) AS AverageAge
FROM Students
GROUP BY City
HAVING AVG(Age) > 21;
```

## Conclusion

The `HAVING` clause is used to filter groups after data has been grouped using the `GROUP BY` clause. It is especially useful when working with aggregate functions like `COUNT()`, `SUM()`, and `AVG()`.
