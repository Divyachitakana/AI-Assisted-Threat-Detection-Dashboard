# SQL – LEFT JOIN

## Explanation

The `LEFT JOIN` clause is used to combine rows from two tables based on a related column. It returns all records from the left table and the matching records from the right table. If there is no matching record in the right table, the result will contain `NULL` values for the columns from the right table.

In this notebook, we will learn how to use the `LEFT JOIN` clause to retrieve all records from one table along with matching data from another table.

## Prompt

> Explain the SQL LEFT JOIN clause in simple language for beginners. Show how to retrieve all records from the left table and matching records from the right table using practical examples.

## Example

```python
-- Students Table
-- StudentID | Name
-- 1         | Alice
-- 2         | Bob
-- 3         | Charlie

-- Courses Table
-- StudentID | Course
-- 1         | Python
-- 2         | SQL
-- 4         | Java

SELECT Students.StudentID,
       Students.Name,
       Courses.Course
FROM Students
LEFT JOIN Courses
ON Students.StudentID = Courses.StudentID;
```

## Conclusion

The `LEFT JOIN` clause returns all records from the left table and the matching records from the right table. If no matching record exists in the right table, SQL displays `NULL` values, making it useful for identifying missing or unmatched data.
