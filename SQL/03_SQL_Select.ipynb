# SQL – SELECT

## Explanation

The `SELECT` statement is one of the most commonly used SQL commands. It is used to retrieve data from one or more tables in a database. You can select all columns or only specific columns based on your requirements.

In this notebook, we will learn how to use the `SELECT` statement to view data stored in a table.

## Prompt

> Explain the SQL SELECT statement in simple language for beginners. Show how to retrieve all columns and specific columns from a table using practical examples.

## Example

```python
-- Select all columns from the Students table

SELECT * FROM Students;

-- Select specific columns

SELECT Name, Age
FROM Students;
```

## Conclusion

The `SELECT` statement is used to retrieve data from a database table. It allows you to view either all columns or only the columns you need, making it one of the most essential SQL commands.
