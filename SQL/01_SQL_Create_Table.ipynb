# SQL – CREATE TABLE

## Explanation

The `CREATE TABLE` statement is used to create a new table in a database. A table consists of rows and columns, where each column stores a specific type of data such as numbers or text.

Before we can insert or retrieve data, we must first create a table that defines the structure of the database.

## Prompt

> Explain the SQL CREATE TABLE statement in simple language for beginners. Show how to create a table with different column names and data types using a practical example.

## Example

```python
CREATE TABLE Student (
    StudentID INT,
    Name VARCHAR(50),
    Age INT,
    City VARCHAR(30)
);
```

## Conclusion

The `CREATE TABLE` statement is used to define the structure of a table before storing any data. Every SQL database starts by creating tables with appropriate columns and data types.
