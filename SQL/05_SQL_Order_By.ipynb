# SQL – ORDER BY

## Explanation

The `ORDER BY` clause is used to sort the records returned by a query. By default, the data is sorted in ascending (`ASC`) order, but you can also sort it in descending (`DESC`) order.

In this notebook, we will learn how to sort data using one or more columns.

## Prompt

> Explain the SQL ORDER BY clause in simple language for beginners. Show how to sort data in ascending and descending order with practical examples.

## Example

```python
-- Sort students by age in ascending order

SELECT *
FROM Students
ORDER BY Age;

-- Sort students by age in descending order

SELECT *
FROM Students
ORDER BY Age DESC;

-- Sort students alphabetically by name

SELECT *
FROM Students
ORDER BY Name ASC;
```

## Conclusion

The `ORDER BY` clause is used to arrange query results in ascending or descending order. Sorting data makes it easier to analyze, compare, and understand information in a database.
