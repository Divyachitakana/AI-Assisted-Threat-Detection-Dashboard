# SQL – WHERE

## Explanation

The `WHERE` clause is used to filter records based on a specific condition. Instead of retrieving all the data from a table, you can use `WHERE` to display only the rows that match your criteria.

In this notebook, we will learn how to use the `WHERE` clause with different conditions.

## Prompt

> Explain the SQL WHERE clause in simple language for beginners. Show how to filter records using comparison operators with practical examples.

## Example

```python
-- Select students whose age is greater than 20

SELECT *
FROM Students
WHERE Age > 20;

-- Select students from Hyderabad

SELECT *
FROM Students
WHERE City = 'Hyderabad';

-- Select a specific student by ID

SELECT *
FROM Students
WHERE StudentID = 1;
```

## Conclusion

The `WHERE` clause helps filter data based on specific conditions. It allows you to retrieve only the records you need, making data searching more efficient.
