# SQL – INSERT

## Explanation

The `INSERT INTO` statement is used to add new records (rows) into a table. Each row contains values that correspond to the columns defined in the table.

Before inserting data, the table must already exist in the database.

## Prompt

> Explain the SQL INSERT INTO statement in simple language for beginners. Show how to insert one row and multiple rows into a table using a practical example.

## Example

```python
-- Insert a single row

INSERT INTO Students (StudentID, Name, Age, City)
VALUES (1, 'Alice', 20, 'Hyderabad');

-- Insert multiple rows

INSERT INTO Students (StudentID, Name, Age, City)
VALUES
(2, 'Bob', 22, 'Mumbai'),
(3, 'Charlie', 21, 'Bengaluru'),
(4, 'David', 23, 'Chennai');
```

## Conclusion

The `INSERT INTO` statement is used to add new records to a table. It allows us to insert one or multiple rows, making it an essential command for storing data in a database.
