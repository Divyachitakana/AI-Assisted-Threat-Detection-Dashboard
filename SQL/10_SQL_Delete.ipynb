# SQL – DELETE

## Explanation

The `DELETE` statement is used to remove existing records from a database table. It allows you to delete one or more rows based on a specified condition.

The `WHERE` clause is important when using the `DELETE` statement because it specifies which records should be removed. If the `WHERE` clause is omitted, all records in the table will be deleted, while the table structure remains unchanged.

In this notebook, we will learn how to delete records using simple SQL statements.

## Prompt

> Explain the SQL DELETE statement in simple language for beginners. Show how to delete specific records from a table using practical examples.

## Example

```python
-- Delete a student with StudentID 4

DELETE FROM Students
WHERE StudentID = 4;

-- Delete students from Chennai

DELETE FROM Students
WHERE City = 'Chennai';

-- Delete students whose age is less than 20

DELETE FROM Students
WHERE Age < 20;
```

## Conclusion

The `DELETE` statement removes records from a database table. Always use the `WHERE` clause to delete only the intended records and avoid accidentally removing all the data from the table.
