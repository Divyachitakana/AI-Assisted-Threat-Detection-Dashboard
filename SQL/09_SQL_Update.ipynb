# SQL – UPDATE

## Explanation

The `UPDATE` statement is used to modify existing records in a database table. It allows you to change the values of one or more columns for specific rows.

The `WHERE` clause is usually used with `UPDATE` to specify which records should be modified. If the `WHERE` clause is omitted, all rows in the table will be updated.

In this notebook, we will learn how to update records using simple SQL statements.

## Prompt

> Explain the SQL UPDATE statement in simple language for beginners. Show how to update one or more column values for specific records using practical examples.

## Example

```python
-- Update the city of a specific student

UPDATE Students
SET City = 'Delhi'
WHERE StudentID = 1;

-- Update the age of a specific student

UPDATE Students
SET Age = 21
WHERE Name = 'Alice';

-- Update multiple columns

UPDATE Students
SET Age = 22,
    City = 'Pune'
WHERE StudentID = 2;
```

## Conclusion

The `UPDATE` statement is used to modify existing records in a table. Using the `WHERE` clause ensures that only the intended rows are updated, preventing accidental changes to all records.
