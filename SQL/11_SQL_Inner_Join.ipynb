# SQL – INNER JOIN

## Explanation

The `INNER JOIN` clause is used to combine rows from two or more tables based on a related column. It returns only the records that have matching values in both tables.

For example, if one table contains student information and another table contains course information, an `INNER JOIN` can be used to display only the students who are enrolled in a course.

In this notebook, we will learn how to combine data from multiple tables using the `INNER JOIN` clause.

## Prompt

> Explain the SQL INNER JOIN clause in simple language for beginners. Show how to retrieve matching records from two tables using practical examples.

## Example

```python
-- Students Table
-- StudentID | Name
-- 1         | Alice
-- 2         | Bob
-- 3         | Charlie

-- Courses Table
-- StudentID | Course
-- 1         | Python
-- 2         | SQL
-- 4         | Java

SELECT Students.StudentID,
       Students.Name,
       Courses.Course
FROM Students
INNER JOIN Courses
ON Students.StudentID = Courses.StudentID;
```

## Conclusion

The `INNER JOIN` clause combines data from two tables by matching related columns. It returns only the rows that exist in both tables, making it useful for retrieving related information stored across multiple tables.
